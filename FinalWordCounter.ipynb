```python
import os
import nltk
from nltk.corpus import stopwords
```

```python
#Load the BigTextFile or Book
book = open('/Users/Samuel/Google Drive/Job/Iprice/plot.list.txt',encoding = "latin-1").read()
```

```python
#split the Book into rows
rows = book.split('\n')
```

```python
# Convert all the words to lowercase and split them
words = [s.lower().split() for s in rows if s]
# Convert the list into a flat list
words = [sublist for l in words for sublist in l]
# Import the stopwords from nltk corpus since we want to include only essintial words that add value not noise
stop = set(stopwords.words('english'))
```

```python
# Remove punctuation and numbers from words
words = [''.join(c for c in w if c.isalpha()) for w in words]
# Remove stopwords and blanks spaces from words
words = [w for w in words if w not in stop and w.isalpha()]
```

```python
# Load the word list into NLTK
text = nltk.Text(words)
# Use NLTK to calculate Frequency distribution
freq = nltk.FreqDist(text)
```

```python
# Print and plot most common words
freq.most_common(30)
freq.plot(15)
```

```python
#As you can see pl,mv, and anonymous are way too common and dont really add value so I will Remove them by adding them to the stopwords set..
stop.add("pl")
stop.add("mv")
stop.add("anonymous")
```

```python
# Now will rerun the last steps and produce a better list of the most frequent words in the book
words = [''.join(c for c in w if c.isalpha()) for w in words]
words = [w for w in words if w not in stop and w.isalpha()]
text = nltk.Text(words)
freq = nltk.FreqDist(text)
freq.most_common(50)
freq.plot(25)
```
